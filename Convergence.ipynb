### Summary

* For weak-strong model the number of electron macroparticles does not affect convergence of the results strongly
* The number of ion macroparicles should simply not be limited
* There is an error associated with changing the macroparticle charge and ion clearing. If there's no clearing this error will dissapear
* Results of weak-weak model vary significantly from results of a weak-strong model.
* Weak-weak model will have a higher total number of macroparticles even to converge

###
The parameters of the simulations for weak-strong model should be:
* $n_\text{segments} = 20$ or lower
* $n_\text{electron macroparticles} = 5\cdot 10^{3}$
* $n_\text{max ion macroparticles} = 2\cdot 10^{4}$, in general this should be unlimited
